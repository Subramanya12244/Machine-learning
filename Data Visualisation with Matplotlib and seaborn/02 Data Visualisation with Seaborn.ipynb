Basic ploting with seaborn


```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
# create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Scatter plot of total bill vs tip ')
```

```python
# Line plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line plot of total bill by size")
```

Categorial plots

Bar plot

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot of total Bill')
```

Box plot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('Box plot of the day vs total bill')
```

violin plot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('violin plot of day vs the total bill plot')
```

Histogram

```python
sns.histplot(tips['total_bill'],bins=10)
# want to remove the curve we use the kde=False if we want twe make it as true
```

KDE plot

```python
sns.kdeplot(tips['total_bill'],fill=True)
```

Pair Plot


```python
sns.pairplot(tips)
```

Heat map

```python
corr=tips[['total_bill','tip','size']].corr()
# in this there should be no categorical column as it wont compare numerical with ccategorical
corr 
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
