```python
import pandas as pd
data=pd.read_csv('googleplaystore.csv')
data=data.dropna()
data
```

```python
category={}
for name in data['Category'].unique():
    ct=0
    for i in data['Category']:
        if i==name:
            ct+=1
    category[name]=ct

print(category)
    
```

```python
content={}
for name in data['Content Rating'].unique():
    ct=0
    for i in data['Content Rating']:
        if i==name:
            ct+=1
    content[name]=ct

print(content)
    
```
