Handling missing values

```python
data_fill=data.fillna(0)
# it is filled with 0 whereever there in no data

```

filling the missing values with mean of the column

```python
data['fill_null']=data['Rating'].fillna(data['Rating'].mean())
data
```

```python
# Renaming the columns
data=data.rename(columns={'Rating':'App rating'})
data.head(5)
```

```python
# change the datatypes
data['Apprating_new']=data['App rating'].fillna(data['App rating'].mean()).astype(float)
```

```python
data['Apprating_new']=data['App rating'].apply(lambda x:x**2)
data['Apprating_new']
```

Grouping and aggregating data

```python
grouped_mean=data.groupby('Category')['App rating'].mean()
print(grouped_mean)
```

```python
grouped_sum=data.groupby(['Category','Reviews'])['App rating'].sum()
print(grouped_sum)
```

```python
# aggrigate multiple functions
grouped_aggr=data.groupby('Category')['App rating'].agg(['mean','sum','count'])
grouped_aggr
```

Merging and joining the DataFrame

```python
df1=pd.DataFrame({'key':['A', 'B', 'C'], 'value':[1,2,3]})
df2=pd.DataFrame({'key':['A', 'B', 'D'], 'value':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
# meging Dataframe on the key columns

pd.merge(df1, df2,on='key',how='inner')
# inner- it will merge the values with same on both sides
```

```python
pd.merge(df1, df2,on='key',how='outer')
# outerr meanss that it will merge the whole table and unmatched values will be kept Nan
```

```python
pd.merge(df1, df2,on='key',how='left')
# left gives more priority to the left table and takes only left table values
```

```python
pd.merge(df1, df2,on='key',how='right')
# right gives more priority to the right table and takes only right table values
```
